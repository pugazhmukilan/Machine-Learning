```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

IMPORTING DATASET

```python
#importing the datset from the storage
dataset = pd.read_csv('Mall_Customers.csv')
x = dataset.iloc[:, [3,4]].values#for visualising the date in the plot we need
#we need the x and y axis so we are taking only the index 3 and 4 for the x parameter

```

USING ELBOW METHOD TO FIND THE OPTIMAL NUMBER OF CLUSTERS

```python
from sklearn.cluster import KMeans
wcss = []
for i in range (1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++',random_state = 42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)#distance between the centroid and the data points


plt.plot(range(1,11),wcss)
plt.title("wcss ELBOW graph")
plt.xlabel("no.of clusters")
plt.ylabel("wcss")
plt.show()
    
```

TRANINING OIF THE K MEANS MODEL ON THE DATASET

```python
#optimal no of clusters is 5 
kmeans = KMeans(n_clusters=5,random_state=42,init = 'k-means++')
y_kmeans = kmeans.fit_predict(x)#this will fit the mdoel and evebnt show which category they fall into
print(y_kmeans)
```

VISUALISING THE CLUSTERS

```python
plt.scatter(x[y_kmeans==0,0],x[y_kmeans ==0 ,1],s = 100,color = 'red',label = 'cluster 1')
plt.scatter(x[y_kmeans==1,0],x[y_kmeans ==1 ,1],s = 100,color = 'green',label = 'cluster 2')
plt.scatter(x[y_kmeans==2,0],x[y_kmeans ==2 ,1],s = 100,color = 'blue',label = 'cluster 3')
plt.scatter(x[y_kmeans==3,0],x[y_kmeans ==3 ,1],s = 100,color = 'orange',label = 'cluster 4')
plt.scatter(x[y_kmeans==4,0],x[y_kmeans ==4 ,1],s = 100,color = 'pink',label = 'cluster 5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color = 'balck',s = 200)
plt.show()
```
