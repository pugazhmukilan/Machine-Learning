```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

IMPORTING DATASET

```python
dataset = pd.read_csv("Mall_Customers.csv")
x = dataset.iloc[:,[2,3]].values
```

CREATING DENDOGRAM


```python
import scipy.cluster.hierarchy as sch

dendogram = sch.dendrogram(sch.linkage(x,method = 'ward',metric ='euclidean'))  # Creating the dend
plt.title("dendogram")
plt.xlabel('customers')
plt.ylabel('eucledian distance')
```

TRAINING THE MODEL

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=4)
y_hc = hc.fit_predict(x)
print(y_hc)
```

VISUALIZING IN THE SCATTER PLOT

```python
plt.scatter(x[y_hc ==0,0],x[y_hc ==0,1],s = 100,color = 'green')
plt.scatter(x[y_hc ==1,0],x[y_hc ==1,1],s = 100,color = 'red')
plt.scatter(x[y_hc ==2,0],x[y_hc ==2,1],s = 100,color = 'blue')
plt.scatter(x[y_hc ==3,0],x[y_hc ==3,1],s = 100,color = 'pink')
plt.legend()
plt.show()
```
